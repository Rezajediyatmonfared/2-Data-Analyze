# سوال 3

```python
data = np.genfromtxt('titanic.csv', delimiter=',', dtype=None, names=True, encoding=None)
# ستون های عددی
numerics = ['Survived','Pclass','Age','SibSp','Parch','Fare']

# چاپ آمار هر ستون
for col in numerics:
    col_data = data[col]

    print(f"\n--- {col} ---")
    print("count =", len(col_data))
    print("mean  =", np.mean(col_data))
    print("std   =", np.std(col_data))
    print("min   =", np.min(col_data))
    print("25%   =", np.percentile(col_data, 25))
    print("50%   =", np.median(col_data))
    print("75%   =", np.percentile(col_data, 75))
    print("max   =", np.max(col_data))
```

```python
# محاسبه میانگین
meanfare = np.mean(data['Fare'])  

# جایگزینی مقادیر صفر با میانگین
fare = data['Fare']
fare = np.where(fare == 0, meanfare, fare)

# چاپ نتایج
print(fare)  
```

# سوال 5

```python
# استخراج ستون‌های Survived و Sex
survived = data['Survived']
sex = data['Sex']

# محاسبه تعداد کل مردان و زنان
total_males = np.sum(sex == 1)
total_females = np.sum(sex == 0)
print(" کل مردان :",total_males)
print("کل زنان :",total_females)
# محاسبه تعداد مردان و زنان نجات‌یافته
male_survivors = np.sum((sex == 1) & (survived == 1))
female_survivors = np.sum((sex == 0) & (survived == 1))
print("مردان نجات یافته :",male_survivors)
print("زنان نجات یافته :",female_survivors)
# محاسبه درصد نجات برای هر جنس
male_survival_rate = (male_survivors / total_males) * 100
female_survival_rate = (female_survivors / total_females) * 100

# چاپ نتایج
print(f"درصد مسافران مرد نجات‌یافته: {male_survival_rate:.2f}%")
print(f"درصد مسافران زن نجات‌یافته: {female_survival_rate:.2f}%")
```

# سوال 6

```python
import numpy as np

# خواندن داده‌ها از فایل
data = np.genfromtxt('titanic.csv', delimiter=',', skip_header=1)

# ستون جنسیت (ستون 4 - اندیس 4)
sex_column = data[:, 4]

# جدا کردن مردان (مقدار 1)
men = data[sex_column == 1]

# جدا کردن زنان (مقدار 0)  
women = data[sex_column == 0]

# اضافه کردن شناسه ترتیبی به مردان
men_with_id = np.column_stack((np.arange(1, len(men) + 1), men))

# اضافه کردن شناسه ترتیبی به زنان
women_with_id = np.column_stack((np.arange(1, len(women) + 1), women))

print("مردان با شناسه ترتیبی:")
print(men_with_id[:5])  # نمایش 5 سطر اول

print("\nزنان با شناسه ترتیبی:")
print(women_with_id[:5])  # نمایش 5 سطر اول
```

# سوال 11

```python
import numpy as np
import matplotlib.pyplot as plt

data = np.genfromtxt('titanic.csv', delimiter=',', skip_header=1)

#ساخت ستون جدید 
family_size = data[:, 5] + data[:, 6]
family_size = np.round(family_size).astype(int)

# شمردن اعداد صحیح 
counts = np.bincount(family_size)
numbers = np.arange(len(counts))

# رسم نمودار
plt.bar(numbers, counts)
plt.title('Passenger according to Family')
plt.xlabel('Family QTY')
plt.ylabel('passenger QTY')
plt.show()
```

# سوال 12

```python
# گرفتن ستون سن (Age) و تبدیلش به عدد
ages = data[:, 5].astype(int)

# بدست آوردن ترتیب مرتب‌شده‌ی سن‌ها (اندیس‌ها)
sorted_index = np.argsort(ages)

# مرتب کردن کل آرایه براساس اندیس‌های بدست آمده
sorted_data = data[sorted_index]

# نمایش چند سطر اول برای بررسی نتیجه
print("۵ مسافر اول بر اساس کم‌ترین سن:")
print(sorted_data[:5])
```

# سوال 14

```python
import numpy as np
import matplotlib.pyplot as plt

# خواندن داده‌ها
data = np.genfromtxt('titanic.csv', delimiter=',', skip_header=1)

# انتخاب دو نفر به صورت تصادفی
i1, i2 = np.random.randint(0, data.shape[0], 2)

# مسافر اول
age1 = data[i1, 5]
comp1 = data[i1, 6] + data[i1, 7]
fare1 = data[i1, 9]

# مسافر دوم
age2 = data[i2, 5]
comp2 = data[i2, 6] + data[i2, 7]
fare2 = data[i2, 9]

# مقادیر برای نمودار
v1 = [age1, comp1, fare1]
v2 = [age2, comp2, fare2]

labels = ['Age', 'Companions', 'Fare']

plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
plt.pie(v1, labels=labels)
plt.title('Passenger 1')

plt.subplot(1,2,2)
plt.pie(v2, labels=labels)
plt.title('Passenger 2')

plt.show()
```
