## Read the Titanic dataset using NumPy.

```python
import numpy as np
import matplotlib.pyplot as plt
data = np.genfromtxt('titanic.csv', delimiter=',', names=True, dtype=float)

print("Data file loaded successfully")
```

## Print the first 10 rows and the last 10 rows

```python
print("First 10 rows:")
for row in data[:10]:
    print(row)

print("\nLast 10 rows:")
for row in data[-10:]:
    print(row)
```

## Calculate statistical features for each numeric column

```python
data = np.genfromtxt('titanic.csv', delimiter=',', dtype=None, names=True, encoding=None)

numerics = ['Survived','Pclass','Age','SibSp','Parch','Fare']


for col in numerics:
    col_data = data[col]

    print(f"\n--- {col} ---")
    print("count =", len(col_data))
    print("mean  =", np.mean(col_data))
    print("std   =", np.std(col_data))
    print("min   =", np.min(col_data))
    print("25%   =", np.percentile(col_data, 25))
    print("50%   =", np.median(col_data))
    print("75%   =", np.percentile(col_data, 75))
    print("max   =", np.max(col_data))
```

## Replace zero values in each numeric column with the mean of that column

```python
meanfare = np.mean(data['Fare'])  

fare = data['Fare']
fare = np.where(fare == 0, meanfare, fare)

print(fare)  
```

## Calculate the survival percentage of male and female passengers

```python
survived = data['Survived']
sex = data['Sex']

total_males = np.sum(sex == 1)
total_females = np.sum(sex == 0)
print("Total males:", total_males)
print("Total females:", total_females)

male_survivors = np.sum((sex == 1) & (survived == 1))
female_survivors = np.sum((sex == 0) & (survived == 1))
print("Male survivors:", male_survivors)
print("Female survivors:", female_survivors)

male_survival_rate = (male_survivors / total_males) * 100
female_survival_rate = (female_survivors / total_females) * 100

print(f"Percentage of male survivors: {male_survival_rate:.2f}%")
print(f"Percentage of female survivors: {female_survival_rate:.2f}%")
```

## Split the array into two separate arrays for men and women, and add a sequential ID to each array

```python
sex_column = data['Sex']

men = data[sex_column == 1]
women = data[sex_column == 0]

print("Males with sequential ID:")
for i, row in enumerate(men[:5], start=1):
    print(i, row)

print("\nFemales with sequential ID:")
for i, row in enumerate(women[:5], start=1):
    print(i, row)
```

## Extract the array of survivors

```python
survived = data[data['Survived'] == 1]

print("Array of survived passengers:")
print(survived[:5])

print("\nNumber of survived passengers:", len(survived))
```

## Count passengers by the Embarked column

```python
emb = data['Embarked']

spain = np.sum(emb == 333)
england = np.sum(emb == 222)
germany = np.sum(emb == 111)

print("Spain:", spain)
print("England:", england)
print("Germany:", germany)
```

## Find the passenger with the highest fare and display their ticket number

```python
fares = data['Fare'].astype(int)
max_fare = np.max(fares)
idx = np.argmax(fares)

print("Row index:", idx)

name = data['Name'][idx]
ticket = data['Ticket'][idx]

print(f"{name}: Passenger")
print("Ticket number:", ticket)
print("Maximum ticket price:", max_fare)
```

## Create a new column by summing Parch and SibSp

```python
sibsp = data['SibSp'].astype(int)
parch = data['Parch'].astype(int)

family_total = sibsp + parch

print("First 5 rows with total family members:")
for i in range(5):
    print(f"Row {i}: {data[i]}, Total Family: {family_total[i]}")
```

## Draw a bar chart showing the frequency of the new column values

```python
family_size = data['SibSp'] + data['Parch']
family_size = np.round(family_size).astype(int)

counts = np.bincount(family_size)
numbers = np.arange(len(counts))

plt.bar(numbers, counts)
plt.title('Passenger distribution according to family size')
plt.xlabel('Number of family members')
plt.ylabel('Number of passengers')
plt.show()
```

## Sort the array by age

```python
ages = data['Age'].astype(float) 

sorted_index = np.argsort(ages)

sorted_data = data[sorted_index]

print("Top 5 passengers with the lowest ages:")
print(sorted_data[:5])
```

## Calculate the average age of male passengers

```python
ages = data['Age'].astype(float)
sex = data['Sex'].astype(int)

male_ages = ages[sex == 1]

mean_male_age = np.mean(male_ages)

print("Mean age of males:", mean_male_age)
```

## Randomly select two passengers and compare their age, number of companions, and travel fare using a pie chart

```python
num_rows = data.shape[0]

i1, i2 = np.random.randint(0, num_rows, 2)

age1 = data['Age'][i1]
comp1 = data['SibSp'][i1] + data['Parch'][i1]
fare1 = data['Fare'][i1]

age2 = data['Age'][i2]
comp2 = data['SibSp'][i2] + data['Parch'][i2]
fare2 = data['Fare'][i2]

v1 = [age1, comp1, fare1]
v2 = [age2, comp2, fare2]

labels = ['Age', 'Companions', 'Fare']

plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.pie(v1, labels=labels, autopct='%1.1f%%')
plt.title(f'Passenger {i1}')

plt.subplot(1, 2, 2)
plt.pie(v2, labels=labels, autopct='%1.1f%%')
plt.title(f'Passenger {i2}')

plt.show()
```
